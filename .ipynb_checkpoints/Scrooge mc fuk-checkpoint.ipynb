```r
# PART 2 OF THE REPORT; GENERALISATION OF THE MODELLING

# Fits the linear regression model and outputs graphs. 25% is not trained date 50% is fitted data and 25% is extrapolated data
# MUST BE TRUE: trend_fit_length >= oscillations_fit_lenght !!!
# Output types are: all, summary, only_full_fit, none
fit_model <- function(fit_end, trend_fit_length, oscillations_fit_length, linear = TRUE, graphs = "summary") {   
    # If negative oscillations fit length is inputted, we assume the user wants no oscillations
    fit_oscillations <- TRUE
    if(oscillations_fit_length < 0) {
        fit_oscillations <- FALSE
    }
    
    # Find length of the fitting period and the total graph xaxis length in seconds
    fit_period_length <- max(trend_fit_length, oscillations_fit_length)
    graph_xaxis_length <- 2*fit_period_length # fit length is 50% of graph
    
    # Find starting date time of trend fitting and oscillations fitting
    fit_end_datetime_obj <- as.POSIXct(fit_end, format = "%Y-%m-%d %H:%M:%S", tz = "UTC")
    
    # Dates for start and ends of graph
    graph_start <- fit_end_datetime_obj - graph_xaxis_length * 0.75 
    graph_end <- fit_end_datetime_obj + graph_xaxis_length * 0.25 
    # Dates for the start of trend fitting and start of oscillation fitting
    trend_fit_start <- fit_end_datetime_obj - trend_fit_length
    oscillations_fit_start <- fit_end_datetime_obj - oscillations_fit_length
    
    # Get the data for the graph, trend fitting and oscillation fitting
    data_graph <- data[data$datetime >= as.POSIXct(graph_start, tz = "UTC") & data$datetime <= as.POSIXct(graph_end, tz = "UTC"), ]
    data_trend_fit <- data[data$datetime >= as.POSIXct(trend_fit_start, tz = "UTC") & data$datetime <= fit_end_datetime_obj, ]
    data_oscillations_fit <- data[data$datetime >= as.POSIXct(oscillations_fit_start, tz = "UTC") & data$datetime <= fit_end_datetime_obj, ]

    # Nicify data
    data_graph$t <- as.numeric(data_graph[['datetime']])
    time_offset <- min(data_graph$t)
    data_graph <- na.omit(data_graph)
    data_graph$t <- data_graph$t - time_offset

    data_trend_fit <- nicify_data(data_trend_fit, time_offset)
    data_oscillations_fit <- nicify_data(data_oscillations_fit, time_offset)
    
    # Create linear model for trend
    lm_trend <- lm(water ~ I(t) + I(t^2), data = data_trend_fit)
    
    # Remove trend from water level for oscillations fitting
    if(nrow(data_trend_fit) >= nrow(data_oscillations_fit)) {
        trend_predictions <- predict(lm_trend, newdata = data_oscillations_fit)
        data_oscillations_fit$water_no_trend <- data_oscillations_fit$water - trend_predictions
    }
    else {
        stop("The oscillation fitting data is bigger than the trend fitting data")
    }

    # Fit the oscillations linear model to predict the oscillations
    if(linear & fit_oscillations) {
        period_1 <- 545
        period_2 <- 205
        period_3 <- 85

        freq_1 <- (2*pi)/period_1 
        freq_2 <- (2*pi)/period_2
        freq_3 <- (2*pi)/period_3 

        model_oscillations <- lm(water_no_trend ~ sin(freq_1*t) + cos(freq_1*t) + sin(freq_2*t) + cos(freq_2*t) + 
                              sin(freq_3*t) + cos(freq_3*t), data = data_oscillations_fit)
    }
    # Fit the oscillations to a non linear model
    else if(fit_oscillations) {
        model_oscillations <- nls(water_no_trend ~ a1*sin(omega1*t) + b1*cos(omega1*t) + a2*sin(omega2*t) + b2*cos(omega2*t) + a3*sin(omega3*t) + b3*cos(omega3*t),
                                control = nls.control(maxiter = 10000, tol = 1e-6),
                                start = list(a1 = 1, b1 = 1, omega1 = .0001, a2 = 1, b2 = 1, omega2 = 0.05, a3 = 1, b3 = 1, omega3 = 1),  
                                data = data_oscillations_fit)
    }   
                               
    # Extrapolate the linear models                         
    data_graph$trend_fit <- predict(lm_trend, newdata = data_graph)
    # If no oscillations should be included
    if(fit_oscillations) {
        data_graph$oscillations_fit <- predict(model_oscillations, newdata = data_graph)
    }
    else {
        data_graph$oscillations_fit <- 0
    }
    data_graph$full_fit <- data_graph$trend_fit + data_graph$oscillations_fit                           
    
    # Date to numeric conversion for graphs
    trend_fit_start_num <- as.numeric(as.POSIXct(trend_fit_start), tz = "UTC") - time_offset
    oscillations_fit_start_num <- as.numeric(as.POSIXct(oscillations_fit_start), tz = "UTC") - time_offset
    fit_end_num <- as.numeric(as.POSIXct(fit_end_datetime_obj), tz = "UTC") - time_offset
                
    # Trend Fitting Graph
    if(graphs == "all") {
        trend_graph_x_len <- 2*trend_fit_length
        graph_trend_start <- fit_end_datetime_obj - trend_graph_x_len * 0.75 
        graph_trend_end <- fit_end_datetime_obj + trend_graph_x_len * 0.25 
        trend_graph <- data[data$datetime >= as.POSIXct(graph_trend_start, tz = "UTC") & data$datetime <= as.POSIXct(graph_trend_end, tz = "UTC"), ]
        trend_graph <- nicify_data(trend_graph, time_offset)
    
        xaxis_labels <- generate_time_labels(trend_graph, num_labels = 6)  
        plot(trend_graph$t, trend_graph$water, type = "l", col = "blue", 
            xlab = "Time", ylab = "Water Level", main = "Trend Fitting Linear Regression Model", xaxt = "n")
        lines(trend_graph$t, predict(lm_trend, trend_graph), col = "red")
        abline(v = trend_fit_start_num, col = "green", lwd = 2, lty = 2)
        abline(v = fit_end_num, col = "purple", lwd = 2, lty = 2)
        legend("topleft", legend = c("Observed", "Fitted", "Fit Data Start", "Fit Data End"), 
            col = c("blue", "red", "green", "purple"), lty = c(1,1,2,2), cex = 0.8)  
        axis(1, at = xaxis_labels$label_positions_t, labels = xaxis_labels$label_times)    
    }
    
    # Water level Without Trend; Oscillations fitting Graph
    if(graphs == "all" & fit_oscillations) {
        xaxis_labels <- generate_time_labels(data_oscillations_fit, num_labels = 6) 
        plot(data_oscillations_fit$t, data_oscillations_fit$water_no_trend, type = "l", col = "blue", 
            xlab = "Time", ylab = "Water Level No Trend", main = "Oscillations No Trend Regression Model", xaxt = "n")
            lines(data_oscillations_fit$t, predict(model_oscillations, data_oscillations_fit), col = "red")
            legend("topleft", legend = c("Observed", "Fitted"), col = c("blue", "red"), lty = 1)
        axis(1, at = xaxis_labels$label_positions_t, labels = xaxis_labels$label_times)
    }
    
    # Plot the raw data against the full fit
    if(graphs == "all" | graphs == "summary" | graphs == "only_full_fit") {
        xaxis_labels <- generate_time_labels(data_graph, num_labels = 6) 
        plot(data_graph$t, data_graph$water, type = "l", col = "blue",
            xlab = "Time", ylab = "Water Level (cm)",
            main = "Raw Data vs Full Model Fit", xaxt = "n")
            lines(data_graph$t, data_graph$full_fit, col = "red")
            abline(v = trend_fit_start_num, col = "green", lwd = 2, lty = 2)
            abline(v = oscillations_fit_start_num, col = "yellow", lwd = 2, lty = 2)
            abline(v = fit_end_num, col = "purple", lwd = 2, lty = 2)
            legend("topleft", legend = c("Raw Data", "Full Model Fit", "Trend Fit Start", "Oscillations Fit Start", "Fit End"),
            col = c("blue", "red", "green", "yellow", "purple"), lty = c(1,1,2,2,2), cex = 0.8)
        axis(1, at = xaxis_labels$label_positions_t, labels = xaxis_labels$label_times)
    }
    
    # Make into arguments maybe?
    interval_len <- 1200
    zoom_range <- data_graph$t >= (fit_end_num - interval_len/2) & data_graph$t <= (fit_end_num + interval_len/2)
    data_zoom <- data_graph[zoom_range, ]
    
    # Plot the zoomed-in raw data and full fit
    if(graphs == "all") {
        xaxis_labels <- generate_time_labels(data_zoom, num_labels = 6)  
        plot(data_zoom$t, data_zoom$water, type = "l", col = "blue",
            xlab = "Time", ylab = "Water Level (cm)",
            main = "Zoomed-in Raw Data and Full Model Fit", xaxt = "n",
            ylim = range(c(data_zoom$water, data_zoom$full_fit)))  # Scale y-axis to fit both lines
            lines(data_zoom$t, data_zoom$full_fit, col = "red")
            abline(v = fit_end_num, col = "purple", lwd = 2, lty = 2)
            legend("topleft", legend = c("Raw Data", "Full Model Fit", "Fit End"),
            col = c("blue", "red", "purple"), lty = c(1,1,2), cex = 0.8) 
        axis(1, at = xaxis_labels$label_positions_t, labels = xaxis_labels$label_times)
    }
    
    # Plot of the 20min error around the fit end
    if(graphs == "all") {
        data_zoom$error <- data_zoom$water - data_zoom$full_fit
        max_error <- max(abs(data_zoom$error), na.rm = TRUE)
        y_range <- range(c(data_zoom$error, -max_error, max_error), na.rm = TRUE)
        xaxis_labels <- generate_time_labels(data_zoom, num_labels = 6)  
        plot(data_zoom$t, data_zoom$error, type = "l", col = "orange",
            xlab = "Time", ylab = "Full Fit Error (cm)",
            main = "Zoomed-in Full Fit Error", xaxt = "n", ylim = y_range)
            abline(v = fit_end_num, col = "purple", lwd = 2, lty = 2)
            abline(h = max_error, col = "red", lwd = 2, lty = 2)  # Positive max error
            abline(h = -max_error, col = "red", lwd = 2, lty = 2) # Negative max error
            legend("topleft", legend = c("Error", "Fit End", paste("Max Abs Error:",round(max_error,1))),
            col = c("orange", "purple","red","red"), lty = c(1,2,2,2), cex = 0.8) 
        axis(1, at = xaxis_labels$label_positions_t, labels = xaxis_labels$label_times)
    }
    
    interval_len_fivemin <- 5*60
    fivemin_range <- data_graph$t >= fit_end_num & data_graph$t <= fit_end_num + interval_len_fivemin
    data_fivemin <- data_graph[fivemin_range, ]
    
    data_fivemin$error <- data_fivemin$water - data_fivemin$full_fit
    max_error_fivemin <- max(abs(data_fivemin$error), na.rm = TRUE)
    
    # 5 minute after fit
    if(graphs == "all" | graphs == "summary") {
        y_range <- range(c(data_fivemin$error, -max_error_fivemin, max_error_fivemin), na.rm = TRUE)
        xaxis_labels <- generate_time_labels(data_fivemin, num_labels = 6)  
        plot(data_fivemin$t, data_fivemin$error, type = "l", col = "orange",
            xlab = "Time", ylab = "Full Fit Error (cm)",
            main = "5 Minute Extrapolation Full Fit Error", xaxt = "n", ylim = y_range)
            abline(h = max_error_fivemin, col = "red", lwd = 2, lty = 2)  # Positive max error
            abline(h = -max_error_fivemin, col = "red", lwd = 2, lty = 2) # Negative max error
            legend("topleft", legend = c("Error", paste("Max Abs Error:",round(max_error_fivemin,1))),
            col = c("orange","red","red"), lty = c(1,2,2,2), cex = 0.8) 
        axis(1, at = xaxis_labels$label_positions_t, labels = xaxis_labels$label_times)    
    }
    
    return(max_error_fivemin)
    
}
   


generate_time_labels <- function(data_stuff, num_labels = 6) {
    time_labels_df <- data.frame(t = data_stuff$t, formatted_datetime = format(data_stuff$datetime, "%H:%M"))
    label_positions_t <- seq(min(data_stuff$t), max(data_stuff$t), length.out = 6)
    label_indices_t <- sapply(label_positions_t, function(pos) which.min(abs(time_labels_df$t - pos)))
    label_times <- time_labels_df$formatted_datetime[label_indices_t]                              
    return(list(label_positions_t = label_positions_t, label_times = label_times))
}
             
                              
                              
nicify_data <- function(data_to_nicify, offset_t) {
    # Convert the datetime field to numeric (seconds)
    data_to_nicify$t <- as.numeric(data_to_nicify[['datetime']])
    # Remove NAs
    data_to_nicify <- na.omit(data_to_nicify)
    #Offset time to 0 to avoid large numbers for t which may cause numerical problems in lm or nls
    data_to_nicify$t <- data_to_nicify$t - offset_t
    return(data_to_nicify)
}
```
